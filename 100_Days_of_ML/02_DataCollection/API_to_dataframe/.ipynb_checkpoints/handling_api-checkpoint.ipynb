# Fetch data from an API

```python
import pandas as pd
import requests
```

```python
# Fetch data from the API and create a DataFrame
url = 'https://api.freeapi.app/api/v1/public/randomusers/user/random'
response = requests.get(url)     # make a GET request to the API
data = response.json()           # translate JSON response to Python data structure
df = pd.DataFrame(data)          # create a DataFrame from the data

# Display the DataFrame
df.head()   
df.info()    # Get information about the DataFrame
df
```

```python
# Extract the information of the random user and creating new dataframe
user_data = df["data"]
# name = pd.DataFrame(user_data["name"])
# gender = pd.DataFrame(user_data["gender"])
# location = pd.DataFrame(user_data["location"])

random_user_df = pd.DataFrame()
user_df = pd.json_normalize(user_data["name"])
location_df = pd.json_normalize(user_data["location"])
random_user_df = pd.concat([user_df, location_df], axis=1)
random_user_df.head()   # Display the new DataFrame with user information

```
