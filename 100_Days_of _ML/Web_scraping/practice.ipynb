```python
import pandas as pd
import requests
from bs4 import BeautifulSoup
import numpy as np
```

```python
webpage = requests.get("https://engineering.careers360.com/colleges/list-of-top-iit-colleges-in-india").text
soup = BeautifulSoup(webpage, "html.parser")
print(soup.prettify())
```

```python
# considering the whole container and extracting the information
college = soup.find_all('div', class_='tupple')

# extracting the details from the main div
college_name = []
ownership = []
location = []
rating = []
nirf = []

for i in college:

    # college name
    name = i.find('h3', class_='college_name').text.strip()
    college_name.append(name)

    # main div
    info = i.find('div', class_='college-listing-wrapper-list')

    # spans
    spans = info.find_all('span')

    # ownership
    ownership.append(spans[0].find('p', class_='college-listing-rating').text.strip())

    # location
    loc_tag = i.find_all('p', class_='college-listing')
    location.append(loc_tag[1].text.strip())

    # rating
    rating_tag = i.find('span', class_='star_text')
    rate = rating_tag.find('b', class_='college-listing-rating').text.strip()
    rating.append(rate)

    # nirf
    nirf_tag = i.find('p', string=' NIRF Ranking:')
    rank = nirf_tag.find_next('b').text.strip()
    nirf.append(rank)

# converting to the dataFrame
df = pd.DataFrame({
    'College Name': college_name,
    'Ownership': ownership,
    'Location': location,
    'Rating': rating,
    'NIRF Ranking': nirf
})

df
```
