# Working with CSV Files in Pandas

## 1. Introduction
Data gathering is a critical step in the machine learning pipeline. While many initial datasets are provided as CSVs (Comma Separated Values), real-world data can come from JSON, SQL databases, APIs, or web scraping. 

**What is a CSV?**
A CSV file stores data where each row is a new line and values are separated by a delimiter, most commonly a comma (`,`).

```python
import pandas as pd
```


## 2. Basic Data Loading

### Loading Local Files
If the CSV is in the same directory as your notebook, you can load it directly by name. Otherwise, provide a relative or absolute path.

```python
df = pd.read_csv('train.csv')
df.head()
```

### Loading from a URL
You can load datasets directly from the web (e.g., GitHub) using the `requests` library to fetch the content before passing it to Pandas.

```python
import requests
from io import StringIO

url = "https://raw.githubusercontent.com/user/repo/main/data.csv"
headers = {"User-Agent": "Mozilla/5.0"}
req = requests.get(url, headers=headers)
data = StringIO(req.text)

df = pd.read_csv(data)
```





## 3. Essential Parameters for `read_csv()`

The `read_csv()` function includes numerous parameters to handle non-standard file formats.

### The `sep` Parameter (Handling TSVs)
If a file uses tabs instead of commas (Tab-Separated Values), use the `sep` parameter with `\t`.

```python
# Loading a Tab Separated Value (TSV) file
pd.read_csv('movie_titles.tsv', sep='\t')
```

### The `names` Parameter
Use this when the CSV file **does not have a header row**. You can pass a list of column names manually.

```python
names = ['sno', 'movie_name', 'release_year', 'rating', 'votes', 'genres']
pd.read_csv('movie_titles.tsv', sep='\t', names=names)
```

### The `header` Parameter
If the column names are accidentally treated as a data row (e.g., row 0 is actually the header), you can explicitly set the header index.

```python
pd.read_csv('data.csv', header=1)
```

### The `usecols` Parameter
To save memory, load only the specific columns you need for your analysis.

```python
pd.read_csv('train.csv', usecols=['gender', 'age', 'target'])
```

### The `index_col` Parameter
If a specific column (like an ID) should serve as the index of the DataFrame, specify it here.

```python
pd.read_csv('train.csv', index_col='enrollee_id')
```






## 4. Advanced Data Handling

### Handling Encodings
If you encounter a `UnicodeDecodeError` (common with emojis or special characters), the file might not be encoded in the default `utf-8`. Try `latin-1` or other specific encodings.

```python
pd.read_csv('spam.csv', encoding='latin-1')
```

### Skipping Bad Lines
In some datasets, a few rows might have more columns than expected, causing parsing errors. Use `on_bad_lines='skip'` (or `error_bad_lines=False` in older versions) to ignore them.

```python
pd.read_csv('books.csv', sep=';', on_bad_lines='skip')
```

### Changing Data Types (`dtype`)
You can reduce memory usage by forcing columns into specific types (e.g., converting a float that only contains whole numbers into an integer).

```python
pd.read_csv('train.csv', dtype={'target': int})
```

### Handling Dates
Pandas often loads dates as strings (objects). Use `parse_dates` to convert them into true datetime objects.

```python
pd.read_csv('ipl_matches.csv', parse_dates=['date'])
```





## 5. Cleaning & Transformation During Load

### Converters
You can apply functions to specific columns while the data is being loaded. For example, shortening team names in a sports dataset.

```python
def rename_team(name):
    if name == "Royal Challengers Bangalore":
        return "RCB"
    return name

pd.read_csv('matches.csv', converters={'team1': rename_team})
```

### Handling Custom Missing Values (`na_values`)
If your dataset uses specific symbols (like `-` or `??`) for missing data, you can tell Pandas to treat them as `NaN`.

```python
pd.read_csv('train.csv', na_values=['-', '??'])
```



## 6. Processing Large Datasets (Chunking)

```python
dfs = pd.read_csv('massive_data.csv', chunksize=5000)

for chunk in dfs:
    # Process each 5000-row chunk individually
    print(chunk.shape)
```
