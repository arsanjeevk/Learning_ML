# Fetch data from an API

```python
import pandas as pd
import requests
```

```python
url = "https://api.freeapi.app/api/v1/public/randomusers/user/random"

# Request
response = requests.get(url)

# Convert to dictionary
data = response.json()

# Extract actual user data
user = data["data"]

# Flatten nested JSON
df = pd.json_normalize(user)

# Display
df.head()
```
